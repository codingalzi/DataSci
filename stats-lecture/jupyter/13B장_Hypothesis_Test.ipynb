#**13장. 가설검정 (2부)**
##1. **가설검정 예제(두번째)**
<u>또 하나의 가설검정을 하고자 한다</u>.    
✏️ **예제**) 코골이 약을 대상으로 이번에는 표본의 크기를 100으로 정함    
**결과**)   

|치료여부|예|아니오|
|:------:|:-:|:---:|
|도수|80|20|

이제 가설검정 6단계 과정을 통하여 치료율이 90%라는 제약회사의 주장을 검정하고자 한다.

1) **단계 1: 영가설과 대립가설 설정**
> **영가설**($H_0$) : $P=0.9 _{(P\ =\ 치료율)}$      
> **대립가설**($H_1$) : $P<0.9$

2) **단계 2 : 검정통계 설정**     
<u>**영가설**이 참</u>이라고 가정할때, 코골이 환자에게 코골이 약이 효과가 있는지 여부는 **이항분포**를 따른다.    
*이는 마치 동전을 던졌을 때 앞면이 나올 확률을 따지는 것과 동일하다.*    
100명을 대상으로 할때, 치료되는 사람의 수를 $X$라 하면 $X$는 아래 모양의 이항분포를 따른다.
$$X ∼ B(100, 0.9)$$
그런데 이 경우 $X$는 정규분포를 근사적으로 따른다.    
이유는 $100 ⋅0.9=90>5$이고 $100⋅0.1=10>5$이기 때문이다.
$$즉,\ X ∼N(90,9)이\ 성립$$
$$_{∵\ 100 ⋅0.9=90}$$
$$_{∵\ 100 ⋅0.9⋅0.1=9}$$

3) **단계 3 : 기각역 설정**
- 유의수준을 5%로 정함
- 기각역은 아래 공식을 만족하는 $C$값들의 집합임
$$P(X<C)<0.05$$
$$하단측검정$$

4) **단계 4 : P값 확인**     
검정통계는 아래 모양의 **정규분포**이다.    
$$X ∼ N(90,9)$$
그리고 표본조사의 경우, <u>100명에 80명이 치료</u>되었다.   
따라서, 80명 이하로 치료될 확률이 **P-값**이 된다.
$$즉,\ P(X ≤ 80) → 표준화 = P(Z ≤ \frac{80-90}{3})=P(Z≤-3.3333)=0.0004$$
> ⬆️ 주의!     
*연속성보정을 가설검정에서는 굳이 하지 않음*

5) **단계 5 : 기각역 포함 여부**    
앞서 구한 **P-값**은 0.0004이며 0.05보다 작다.    
따라서 **80명이란 숫자는 기각역에 포함**된다.

6) **단계 6 : 최종 결정**    
100명 중에 치료된 사람의 수인 80이 기각역에 포함된다.    
따라서 <u>**치료율이 90%라는 영가설을 기각**</u>한다.

##2. **가설검정 오류와 검정력**
- 통계적 주장(가설)은 전혀 절대적이지 않으며 확률적으로만 의의를 갖는다.
- 그러한 가설을 검정한 결과 또한 절대적이지 않으며, 확률적으로만 의의를 갖는다.

✏️ **예제**) 코골이 치료제의 경우    
1. 치료율이 90%라는 영가설($H_{0}$)이 참인 경우       
  (1) 검정결과 $H_0$를 채택하는 경우 : **옳은 결정**       
  (2) 검정결과 $H_0$를 기각하는 경우 : **제 1종 오류 발생**      
2. 치료율이 90%라는 영가설($H_0$)이 거짓인 경우      
  (1) 검정결과 $H_0$를 채택하는 경우 : **제 2종 오류 발생**       
  (2) 검정결과 $H_0$를 기각하는 경우 : **옳은 결정** *("검정력" 입증)*      

> 주의!        
*폭스바겐 스캔들의 경우는 2-(2)에 해당함*     
*즉, 가설 검정의 검정력이 입증된 경우임*

**위 네 가지 경우를 정리하면 아래 표와 같다**

|   |$H_0$를 채택|$H_0$를 기각|
|:--:|:----------:|:----------:|
|$H_0$이 참|옳은 결정|제 1종 오류|
|$H_0$이 거짓|제 2종 오류| 검정력 입증|

- **제 1종 오류가 발생할 확률**
  - 유의수준과 같음
  - 영가설이 예를 들어 **95%의 신뢰수준**으로 `참`이기 때문에,    
  그렇지 않은 경우, 즉 **영가설을 기각할 확률**이 나머지 `5%`임

- **제 2종 오류가 발생할 확률**
  - 보다 복잡한 계산 과정이 요구됨
  - 여기서는 제 2종 오류에 대한 감각만을 익히면 됨

- **검정력**
  - **영가설이 거짓**인데 <u>그 사실을 맞출 확률</u>
  - $1 - (제\ 2종\ 오류\ 발생\ 확률)$
