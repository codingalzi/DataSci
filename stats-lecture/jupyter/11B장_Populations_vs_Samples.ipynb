##3. **모집단의 비율 추정하기**
1. **점추정** : 표본을 통해 구한 정보(평균, 분산 비율 등)를 이용하여 모집단에 대한 정보를 구하는 방법    
⇒ 점추정을 이용하여 구한 모집단에 대한 정보(*평균, 분산 비율 등*)가 정확하다고 장담할 수는 없다.     
하지만 그렇게 하는 것이 최선이다.    
→ *주의 : 편향되지 않은 표본이 구성되도록 최선을 다해야 한다*

2. **모집단 비율의 점추정**
> 1. 표본이 갖는 <u>성공률</u>을 구한다.
$$P_{s} = \frac{성공\ 횟수}{표본\ 크기}$$
> 2. 모집단 성공률의 추정치를 설정한다.
$$ \hat{P} = P_{s} $$
>  $$_{즉,\ 표본의\ 성공률을\ 모집단\ 성공률의\ 추정치로\ 사용}$$
> 3. 확률과 비율의 관계       
> 임의로 선택된 사람에게 대통령에 대한 지지여부를 물어서 긍정적인 답을 얻으면 성공이라고 하였을 때, 모집단에서의 성공의 비율은 대통령의 지지율과 동일한 값이다. 이와 같이, 성공할 확률을 계산하는 것과 성공의 비율을 계산하는 과정은 완전히 동일한다.


✏️ **예제**) "달콤한 풍선껌" 회사에서 개발한 껌에 대한 선호도를 조사하였다.    
40명을 임의로 선택하여 물었더니 그 중 10명이 분홍색 풍선껌을 선호하였다.

```
질문 ❔    
1. 모집단에서 분홍색 풍선껌을 선호하는 사람의 비율은 얼마인가?    
  즉, 사람을 임의로 골랐을 때 그 사람이 분홍색 풍선껌을 선호할 확률은 얼마인가?
2. 모집단에서 분홍색 풍선껌을 좋아하지 않는 사람을 선택한 확률은 얼마인가?
```
>  답안 ❕    
1. $ \hat{P} = P_{s} = \frac{10}{40} = 0.25$
2. $P(분홍색을\ 선호하지\ 않는\ 사람)= 1- \hat{P} = 1-0.25 = 0.75$

##과제 (*추가내용*)
1. 과제 안내    
  11장~12장 내용을 이해하기 위해 다음의 모의실험을 각자 집에서 실행해 볼 것    
  직접 모의실험을 해 본 경우 강의 내용을 보다 쉽게 이해 가능

2. 모의 실험 내용    
  (1) 하나의 주사위가 주어졌다고 가정하자      
  (2) 주어진 주사위를 30번 던졌더니 숫자 1이 10번 나왔다.     
  즉, 숫자 1이 나온 비율이 $\frac{1}{3}$이었다. 

3. 질문     
  (1) 주어진 주사위가 정상적인 주사위인가?      
  (2) 질문 1에 대한 본인의 답에 대한 근거를 설명하라.

4. 힌트      
  (1) 가정 : 주어진 주사위는 정상이다.     
  즉, "6번에 1번의 비율로 숫자 1이 나온다"고 가정      
  그러면 주사위를 30번 던져서 숫자 1이 나오는 횟수 X는 이항분포를 따른다.
  $$X ∼B(30,\frac{1}{6}) $$
  > **30** = 주사위 30번 던지기    
  > $\frac{1}{6}$ = 주사위를 한 번 던졌을 때 숫자 1이 나올 확률

  (2) 따라서, 주사위를 30번 던졌을 때 숫자 1이 나온 횟수의 비율을 나타내는 분포는 $\frac{X}{30}$이다. 즉, $P_{s}$는 아래의 분포를 따른다.
  $$P_{s} = \frac{X}{30}$$
  (3) 이제 $P_{s}$의 기대치, 즉 평균값과 분산을 구할 수 있다.
  $$E(P_{s})=E(\frac{X}{30})=\frac{1}{30}E(X)=\frac{1}{30}E(X)=\frac{1}{30}⋅30⋅\frac{1}{6}=\frac{1}{6}=0.1677$$

   $$Var(P_{s})=Var(\frac{X}{30})=\frac{1}{{30}^2}⋅Var(X)=\frac{1}{{30}^2}⋅30⋅\frac{1}{6}⋅\frac{5}{6}=\frac{1}{{6}^3}=0.0046=(0.068)^2$$

  (4)또한 $P_{s}$는 <u>정규분포를 따른다</u>. $_{←중심극한정리에\ 의해}$
 
  $$P_{s} ∼ N(0.1667, (0.068)^2)$$

  (5) 이제 정상적인 주사위를 던져서 숫자 1이 나오는 비율이 $\frac{1}{3}$ 이상일 확률을 계산할 수 있다.
  $$P(P_{s}≥0.3333)=P(P_{s}>0.333-\frac{1}{2⋅30})_{←연속성\ 보정(추후\ 설명)}$$
  $$P(P_{s}>0.3167)=P(Z>\frac{0.3167-0.1667}{0.068})=P(Z>2.21)=1-P(Z≤2.21)=1-0.9864=0.014$$
  $$(즉,\ 1.4\%)$$

5. 결론    
  주어진 주사위가 정상적이다라는 가정하에서, 주사위를 30번 던져서 숫자 이 나오는 비율이 $\frac{1}{3}$이상일 확률은 1.4%에 불과하다.     
  어떤 가정하에서 발생할 확률이 5% 미만인 사건이 실제로 발생하였다면 해당 가정이 "95% 확률로" 잘못되었다고 결론 내린다.     
  즉, "**주어진 주사위가 정상이다**"라는 가정을 받아들이지 않는 게 일반적이다.

##5. **비율의 표본분포 예제**
✏️**예제**)    
- 풍선껌을 100개씩 상자에 담아 판매
- 모집단에서 빨간색 풍선껌의 비율은 $P=0.25$로 주어짐

```
질문 ❔     
풍선껌 100개들이 한 상자에 빨간색 풍선껌의 비율이 40% 이상일 확률은 ?    
```
$$즉,\ P(P_{s}≥0.4)=?$$

*예제의 질문에 답하기 위해서는 아래 과정을 확인해야 한다.     
→ 풍선껌 100개들이 한 상자를 무작위로 선택하였을 때 빨간색 풍선껌의 **비율에 대한 표본분포**를 확인한다.*

> 답안 ❕     
X를 100개들이 한 상자에 들어 있는 빨간색 풍선껌의 개수라 하자. 그러면 X는 이항분포를 따른다.   
$$X ∼ B(100,0.25)$$   
따라서 100개들이 한 상자에 들어 있는 빨간색 풍선껌의 비율의 분포를 $P_{s}$라 하면 다음이 성립한다.    
$$P_{s} ∼ \frac{X}{100}$$    
$P_{s}$의 <u>평균값과 분산을 다음과 같이 구한다.</u>    

> **평균값**    
$E(P_{s})=E(\frac{X}{100})=\frac{1}{100}E(X)=\frac{1}{100}⋅{100}⋅\frac{1}{4}=0.25$      
> **분산**   
$Var(P_{s})=Var(\frac{X}{100})=\frac{1}{{100}^2}⋅100⋅\frac{1}{4}⋅\frac{3}{4}=\frac{3}{1600}=0.001875$

> 이제, $P(P_{s}≥0.4)$를 구하고자 한다.    
그런데 $P_{s}$는 어떤 분포를 따르는가 ?    
⇒ **정규분포**를 따른다.    
즉, $P_{s} ∼ N(0.25, 0.001875)$가 성립한다. *(이유 : **중심극한정리**)*   
$$P(P_{s}≥0.4)=P(P_s>0.4-\frac{1}{2⋅100}_{연속성\ 보정\ 필요!})=P(P_s>0.395)=P(Z>\frac{0.395-0.25}{\sqrt{0.001875}})=P(Z>3.35)$$
$$=1-P(Z≤3.35)=1-0.9996=0.0004$$
$$_{0.04\% → 사실상\ 가능성\ 없음!}$$

- **연속성 보정**
  - 표본의 크기 $n$이 1,000 이하일 경우 연속성 보정 필요 $(\pm \frac{1}{2n})$

##6. **비율의 표본분포(일반화)**
모집단에서 특정 사건이 발생할 확률이 $P$라고 하자.   
그러면 $n$개의 표본에서 해당 사건이 발생하는 횟수 X는 이항분포를 따른다. 
$$X∼B(n,p)$$
따라서, $n$개의 표본에서 해당 사건이 발생할 확률 $P_s$의 분포는 $P_s=\frac{X}{n}$이다.    
$P_s$의 평균값과 분산은 다음과 같다.
- 평균값
$$E(P_s)=E(\frac{X}{n})=\frac{1}{n}⋅E(X)=\frac{n⋅p}{n}=p$$
- 분산
$$Var(P_s)=Var(\frac{X}{n})=\frac{1}{{n}^2}⋅Var(X)=\frac{1}{{n}^2}⋅n⋅p⋅q=\frac{p⋅q}{n}$$
$$_{(q=1-p)}$$   

또한, $n≥30$일때 $P_s$는 <u>중심극한정리에 의해</u> **정규분포**를 따른다.
$$P_s∼N(P,\frac{p⋅q}{n})$$

---

##7 $P_s$**의 연속성 보정**
> $P_s = \frac{X}{n}$    

> $X = 표본이\ 갖는\ 성공의\ 수$

⇒연속성 보정 값 = $±\frac{1}{2n}$

즉, $P_s$를 위한 확률을 구하기 위해 정규분포를 이용한 근사치를 구하고자 할 때 $±\frac{1}{2n}$이라는 **연속성 보정**을 적용해야 한다.    
$$(예)\ \  P(P_s≥0.40)=P(P_s>0.40-\frac{1}{2⋅100})$$

하지만 $n≥1,000$인 경우 <u>연속성보정 불필요</u>!
