### 1. 점추정의 문제와 해결책

##### 1) 문제 
__①__ 모집단의 평균과 분산을 위해 표본을 활용한 점추정 기법을 살펴보았다. 그런데 추정값을 얻기 위해 단 하나의 표본만을 이용하였다.

__②__ 표본이 편향되지 않았다 하더라도 표본이 모집단을 100% 정확하게 반영하는지 여부는 절대로 알 수 없다.

__③__ 점추정은 단지 최선을 예측하는 것 뿐이다.

##### 2) 해결책
__①__ 예를 들어, 평균값에 대한 추정값을 명확하게 말하는 대신에 평균의 추정값이 어느 구간에 속하는지를 말할 수 있다.

__②__ 구간의 크기는 결과를 어느 정도 신뢰할 수 있게 만들 것인가에 따라 결정된다.

__③__ 신뢰할 수 있는 정도를 __신뢰수준__이라 부름.

### 2. 점추정 신뢰구간

𝟭) __✏️예제__ : 달콤한 풍선껌 회사의 표본조사 결과  
${μ}̂ = x̅=62.7$(분)  
$s^2 = 25$  
$n = 100$ ⬅ 표본의 크기  

$  $

__질문__ :  ${μ}̂$를 어느정도로 신뢰할 수 있나?

#####  

𝟮) __해결책__  
- 모평균 μ가 위치할 가능성이 높은 구간을 ${μ}̂$̂, $s^2$을 활용하여 계산함

𝟯) __준비사항__
- 모평균 μ가 위치할 가능성이 높은 구간을 신뢰구간이라하며 신뢰구간을 계산하기 위해 아래 사항이 준비되어야함.

① <U>표본평균의 분포 확인</U>  
- 모집단을 대상으로 하는 분포가 어떤 분포를 따르는지 모름. 
- 반면에 표본들의 분포는 표분의 크기가 충분히 크면 정규분포 따름.
- 따라서 모평균에 대한 정보를 유추할 수 있음  

② <U>신뢰수준 설정</U>
- 신뢰수준 : 설정한 신뢰구간이 모평균을 어느 정도의 확률로 포함할 지에 대한 기준 제공
- 일반적으로 90%, 95%, 99% 사용하며, 이중에 95% 신뢰수준이 가장 많이 사용됨.
- 신뢰수준을 높일수록 신뢰구간이 넓어짐.
- 신뢰구간이 너무 크면 신뢰구간의 의미가 약해짐

𝟰) __풍선껌 모집단의 모평균 신뢰구간 계산__  
① 평균의 표본분포 확인 
- 풍선껌 향의 지속시간을 가리키는 확률변수를 $X$라 하자.  
이제 모집단의 기대치와 분산을 아래와 같이 점추정할 수 있다.  
${μ}̂=x̅=62.7$  
${σ}̂^2=s^2=25$  
$X$가 어떤 분포를 따르는지 모른다. 따라서 모평균 ${μ}$에 대한 어떤 정보도 유추할 수 없다.  
반면에 표본의 크기가 충분히 크면 $X̅$는 정규분포를 따르며 $X̅$의 기대치와 $X$의 기대치가 동일하다는 사실을 이용하여 모평균 ${μ}$가 포함될 가능성이 높은 신뢰구간을 계산할 수 있다.  
풍선껌 100개로 구성된 표본들에 대한 향 지속시간의 평균값의 분포 $X̅$의 통계량은 다음과 같다.  
$E(X̅)=μ$  
$Var(X̅) = \frac{s^2}{n}=\frac{25}{100}=0.25$
  - $s^2 : σ^2$을 모를 때 대신 사용  
  
  또한 X̅는 정규분포를 따른다. 즉,  
  $X̅$~$N(μ, 0.25)$
  
② 신뢰수준 설정  
- 95%의 신뢰수준 사용

③ 모평균 신뢰구간 계산
- 전제 : 95% 신뢰수준 사용
- 모평균의 신뢰구간 $[a, b]$는 다음을 만족해야한다.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
plt.rc("font", family='NanumGothic')


x = np.arange(-5, 5, 0.01)

def gaussian(x, mean, sigma):
  return (1 / np.sqrt(2 * np.pi * sigma**2)) * np.exp(-(x - mean)**2 / (2 * sigma**2))

fig = plt.figure(figsize=(10, 5))
ax = fig.add_subplot(1, 1, 1)


plt.plot(x, gaussian(x, 0, 1))
plt.fill_between(x, norm.pdf(x,0,1),where=(x>=-1.3)&(x<=1.3), alpha=0.5)
plt.fill_between(x, norm.pdf(x,0,1),color='red', alpha=0.2)

plt.xlabel('μ̂ =x̅')

ax.annotate('25%', xy=(1.7 ,0.02), xytext=(3, 0.1),
            arrowprops=dict(facecolor='red', shrink=0.05))
ax.annotate('25%', xy=(-1.7 ,0.02), xytext=(-3, 0.1),
            arrowprops=dict(facecolor='red', shrink=0.05))
ax.annotate('95%', xy=(0 ,0.3), xytext=(2, 0.25),
            arrowprops=dict(facecolor='violet', shrink=0.05))

plt.show()
```

즉 $P(a<x̅<b)=0.95$를 만족시키는 $a$와 $b$를 구해야 함.  

__방법__ : 표준화 과정 거꾸로 하기

__ⅰ__) $X̅$의 표준화 : $Z = \frac{X̅-μ}{\sqrt{0.25}}$  

__ⅱ__) $P(z_a<z<z_b) = 0.95$가 되도록 $z_a,\ z_b$구하기. 즉, 다음이 성립해야한다.
- $P(z<z_a) = 0.025$
- $P(z>z_b) = 0.025$  
▶ $z_a=-1.96,\ z_b=1.96$

__ⅲ__) $z_a,\ z_b$를 이용하여 $μ$의 신뢰구간 구하기.  
힌트 : 다음 공식 활용  

- $-1.96<z=\frac{X̅-μ}{\sqrt{0.25}}<1.96$
- $X̅-0.98<\mu<X̅+0.98$

__ⅳ__) X̅로 사용 가능한 값 확인 
- X̅는 표본 평균의 분포를 가리킴
- 따라서 X̅의 기대치의 추정값인 표본 평을 x̅를 사용할 수 있음.

__ⅴ__) 모평균 μ의 신뢰구간  
- $[62.7-0.98,\ 62.7+0.98]$  
$= [61.72,\ 63.68]$

### 4. 신뢰구간 설정 공식 모음

__①__
- 모집단 통계 : μ  
- 모집단 분포 : $X$ 정규분포  
- 조건 : 
  - $σ^2$ 알고있음  
  $n$ 표본 크기  
  $X̄$표본 평균 값
- 신뢰구간 : $(X̄-c\frac{σ}{\sqrt{n}},\ X̄+c\frac{σ}{\sqrt{n}})$  

__②__
- 모집단 통계 : μ  
- 모집단 분포 : $X$ 정규분포 아님
- 조건 : 
  - $σ^2$ 알고있음  
  $n \ge 30$  
  $X̄$표본 평균 값
- 신뢰구간 : $(X̄-c\frac{σ}{\sqrt{n}},\ X̄+c\frac{σ}{\sqrt{n}})$  

__③__
- 모집단 통계 : μ  
- 모집단 분포 : $X$ 임의의 분포  
- 조건 : 
  - $σ^2$ 모름  
  $n \ge 30$  
  $X̄$표본 평균 값
  $S^2$표본 분산
- 신뢰구간 : $(X̄-c\frac{s}{\sqrt{n}},\ X̄+c\frac{s}{\sqrt{n}})$  





▶ c의 값 : 신뢰 수준에 의해 결정됨

|신뢰수준|c의 값|
|:---:|:---:|
|90%|1.64|
|95%|1.96|
|99%|2.58|
