|완치 여부|그렇다|아니다|
|:-------:|:----:|:----:|

##3. **가설검정 단계 1**
###<u>1. 영가설과 대립가설</u>
검정할 가설을 설정해야 한다.    
이와 같이, 검정 대상으로 사용되는 주장을 **영가설(Null Hypothesis)**라 부르며 보통 $H_0$라 표기 (*책에 따라 "귀무가설"이라 부르기도 함*)     
반면에 "<u>영가설에 반하는 내용</u>"을 담은 가설은 **대립가설**이라 부른다.

✏️ **예제**) "코골이 약"의 경우    
$영가설(H_0) : P_{(=치료율)}\ \ = 0.9$      
$대립가설(H_1) : P < 0.9$    

> *주의!*     
여기서는 대립가설로 "$P \neq 0.9$"를 사용하지 않는다.


###<u>2.대립가설 설정</u>
1. 앞서 설명한 대로 대립가설은"영가설에 반하는 내용"을 담아야 한다.     
즉, **영가설($H_0$)을 기각하는 경우에 사실로 인정해야 하는 가설이 <u>대립가설</u>이어야 한다.**
2. "코골이 약" 예제를 다시 살펴보자.    
영가설($H_0$)이 "$P=0.9$"이다. 그런데 단순히 수식으로만 생각해서 "$P \neq 0.9$"를 대립가설로 설정하면 문제가 발생한다.    
  > *이유는 "$P \neq 0.9$"에 "$P>0.9$"의 내용이 포함되어 있기 때문이다*.    
  다시 말해서, 제약회사의 주장인 "**치료율 90%**"를 **기각**하면서 <u>치료율이 "90% 이상"이라는 내용을 받아들일 수는 없기 때문</u>이다.      
  따라서 이 경우에는 "$P<0.9$"를 **대립가설**로 설정해야 한다.
3. "다른 예제"를 살펴보자.     
  (1) 어떤 자동차회사에서 생산하는 부품의 "**대리점 판매가의 평균이 15만원**"이라는 주장을 검정하고자 한다.    
  위 주장을 검정하기 위해서 여러 대리점을 돌면서 해당 제품의 가격을 확인해 보고자 한다.     
  이때 사용하는 **영가설**과 **대립가설**은 다음과 같다.
$$H_0 : 𝝁 = 15$$
$$H_1 : 𝝁 >15$$
$$_{(단위\ 만원)}$$
> **$H_1$ 설정 이유**     
"𝝁=15"를 기각할 경우 "𝝁<15"를 포함하는 것은 설정상 바람직하지 않다.     
이유는 주장보다 저렴한 가격에 부품을 구입할 수 있다면 어느 누구도 문제 삼지 않을 것이기 때문이다.

  (2) 어떤 자동차회사에서 "**새로 출시한 차종의 연비가 $15km/l$이다**"라는 주장을 검정하고자 한다.    
  위 주장을 검정하기 위해서 동일 차종 여러 대를 대상으로 연비를 확인해 보고자 한다.    
  이때 사용하는 **영가설**과 **대립가설**은 다음과 같다.
$$H_0 : 𝝁 = 15$$
$$H_1 : 𝝁 <15$$
$$_{(단위\ km/l)}$$    
> **$H_1$ 설정 이유**     
"𝝁=15"를 기각할 경우 "𝝁>15"를 포함하는 것은 설정상 바람직하지 않다.     
이유는 주장보다 높은 연비가 나올 경우 어느 누구도 문제 삼지 않을 것이기 때문이다.

  (3) 어떤 자동차회사에서 "**주식 배당율이 15%이다**"라는 주장을 검정하고자 한다.    
  위 주장을 검정하기 위해서 배당을 받은 여러 주주들을 대상으로 배당율을 확인해 보고자 한다.     
  이때 사용하는 **영가설**과 **대립가설**은 다음과 같다.
$$H_0 : 𝝁 = 15$$
$$H_1 : 𝝁 \neq 15$$
$$_{(단위\ \%)}$$    
> **$H_1$ 설정 이유**     
"𝝁=15"를 기각할 경우 "𝝁>15"와 "𝝁<15"가 모두 포함되어야 한다.     
이유는 "𝝁>15"일 경우 회사에 손해가 보다 크고, "𝝁<15"일 경우 주주들에게 보다 많은 손해가 발생하기 때문이다.

4. 대립가설 설정의 중요성     
대립가설은 "가설 검정의 단계 3"에서 설명할 "기각역 설정"과 밀접하게 연관되어 있다.    
즉, 대립가설을 어떻게 설정하는가에 따라 기각역의 크기와 위치가 달라진다.    
아래에서 설명할 "단계 3"에서 앞서 언급한 자동차 회사의 3개의 예제를 다시 활용하여 각기 다른 기각역 설정 방법을 설명할것이다.

##4. **가설검정 단계 2**
1. 설정된 가성을 검정하는 데 사용할 통계인 **검정통계**를 선택한다.    
즉, <u>검정과정 자체와 관련이 깊은 통계</u>를 설정해야 한다.
2. 코골이 예제인 경우
  - <u>15명으로 구성된 표본에서 치료된 사람의 수에 대한 통계</u>를 선택해야 한다.
  - $X$가 15명 중에서 치료된 이의 수를 나타낸다면 **영가설**($H_0:P=0.9$)에 의해 $X$는 아래 모양의 이항분포를 따른다.    
  $$X∼B(15,0.9)$$
3.주의 사항     
  (1) 영가설($H_0$)이 옳다는 가정하에 결정 통계 선정 → 코골이 예제의 경우는 이항분포 사용     
  (2) 영가설이 틀리다는 강력한 증거가 나오기 전까지 영가설을 받아들여야 함       
  (3) 그런 다음에 **표본의 결과가 발생할 확률**을 계산하여 **기각역에 포함되는지 여부**를 확인한다.

##5. **가설검정 단계 3**
1. **기각역** : 영가설에 극단적으로 반하는 증거에 해당하는 값들의 집합 
2. 코골이 치료약의 경우 "치료율이 90%이다"라는 영가설이 참이라면 15명중 13~14명이 치료되어야 한다.      
하지만 11명만 치료되었는데 이것이 우연의 결과일 수도 있고 아닐 수도 있다.    
반면에, <u>15명 중에 5명 정도면 치료되었다면 제약회사의 주장을 아마 절대로 받아들이지 않을 것</u>이다. 


````python
import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic') 
plt.axhline(y=5, color='r', linewidth=1)
plt.axvline(12, 0.2, 0.8, color='lightgray', linestyle='--', linewidth=2)
plt.gca().axes.yaxis.set_visible(False)
plt.xlim(0,15)
plt.xlabel("명")
plt.annotate('', ha = 'center', va='top', xytext = (11, 5.05), xy = (1, 5.05), arrowprops = {'facecolor':'b', 'edgecolor':'b', 'arrowstyle': '<->'})
plt.annotate('', ha = 'center', xytext = (15, 5.05), xy = (12, 5.05), arrowprops = {'facecolor':'b', 'edgecolor':'b', 'arrowstyle': '<->'})
plt.text(6, 5.1,'H0틀릴 가능성 높아짐',fontsize=10,va='top',ha='right')
plt.text(14, 5.1,'H0 인정',fontsize=10,va='top',ha='right')
plt.show()
````

  그렇다면 15명 몇 명 이하로 치료되면 영가설($H_0$)을 기각할지를 정할 필요가 있다.    
  이와 같이 영가설에 강력히 반하는 값들의 집합을 **기각역**이라 한다.    
  코골이 예제의 경우 15명 중 몇 명 이하로 치료되면 영가설을 기각할 것인가에 판단 기준이 되는 <u>"몇명 이하"에 해당하는 구간이 **기각역**</u>이 된다.

3. 유의수준     
  (1) 기각역을 정하기 위해서는 대립가설과 함께 **유의수준**을 고려한다.       
  (2) **유의수준** : 영가설을 기각해야 한다는 결정을 내릴 때 그 결정이 틀릴 확률.     
  보통 1% 또는 5%를사용하며, 여기서는 5%를 사용하며 설명한다.     
  $X%가 코골이 약으로 치료된 수라 할 때 기각역은 아래 식을 만족하는 $C$ 값들의 집합이다.
  
  $$P(X<C)<0.05$$

  $$X<C인\ 이유\ :\ 대립가설(H_1)이\ 'P<0.9'이기\ 때문$$ 

  (3) 대립가설의 역할     
  앞서 단계 1에서 설명한 대로 동일한 영가설에 대해 **상황에 따라 다른 대립가설을 설정**해야 한다.      
  그리고 어떤 대립가설을 사용하는가에 따라 기각역의 위치와 범위가 달라진다.     
  **코골이 예제의 경우** : $H_1 : P<0.9$     
  앞서 설명한 대로 **기각역**은 <u>몇 명 이하일 때 영가설을 기각할지를 정하는 기준</u>이 되어야 한다.    
  따라서 기각역이 앞서 설명한 대로 아래식을 만족하는 $C$ 값들의 집합이 된다.
  $$P(X<C) <0.05$$
  ⇒ 이와 같이 <u>어떤 값보다 꽤 적게 나올 확률이 **5%**미만인 값</u>들로 이루어진 기각역을 사용하는 방식을 **하단측검정**이라 부른다.    
  ⇒ 반면에, 자동차 부품 대리점 판매가의 경우에는 <u>어떤 값봐 꽤 많이 나올 확률이 **5%**미만인 값</u>들로 이루어진 기각역을 사용해야 하며, 이런 방식을 **상단측검정**이라 한다. 따라서, 상단측검정의 기각역은 아래식을 만족하는 $C$ 값들의 집합이 된다.
  $$P(X>C)<0.05$$
  또한, 자동차 회사 주식배당율의 경우에는 어떤 값과 다를 확률을 다루어야 한다.    
  ⇒ 이럴때에는 <u>주어진 값보다 꽤 크거나, 아니면 꽤 작을 확률이 **각기 2.5%**미만인 값</u>들로 이루어진 기각역을 사용해야 하며, 이런 방식을 **양측 검정**이라 한다.    
  따라서 양측 검정의 기각역은 아래식을 만족하는 $C$ 값들의 집합이 된다.
  $$P(X<C)<0.025\ 또는\ P(X>C)<0.025$$

> *유의수준을 1%로 정할 경우, 단측 검정에서는 한쪽 끝에서 1.1%를 기준으로 기각역 설정*     
*유의수준을 1%로 정할 경우, 양측 검정에서는 양쪽 끝에서 0.5%를 기준으로 기각역 설정*

##6. **가설검정 단계 4**
1. **P-값** : 앞서 단계 2에서 설정한 검정통계를 이용하여 계산된 특정 사건이 발생할 확률
2. <u>코골이 약의 경우</u> : 15명 중에서 11명이 완치될 확률이 **P-값**이 된다.    
⇒ 즉, $X ∼ B(15, 0.9)일\ 때,\ P(X≤11)이\ P-값$

$$P-값 : P(X≤11)=1-P(X≥12) = 1-(P(X=12)+P(X=13)+P(X=14)+P(X=15))$$

$$=1-(^{15}C_{12}⋅(0.9)^{12}⋅(0.1)^{3}+^{15}C_{13}⋅(0.9)^{13}⋅(0.1)^{2}+^{15}C_{14}⋅(0.9)^{14}⋅(0.1)^{1}+^{15}C_{15}⋅(0.9)^{15})$$

$$=0.0555 >0.05$$

##7. **가설검정 단계 5**
1. 단계 4에서 구한 **P-값**이 <u>기각역 안에 포함되는지를 확인</u>한다.
2. 코골이 약의 경우, P-값이 0.0555이며, 0.05보다 큰 값이므로, **치료된 사람의 수인 11은 기각역에 포함되지 않는다.**

1. 단계 5의 결과에 따라 <u>영가설에 대한 **기각여부**를 판단</u>한다.      
  (1) 기각역에 포함됨 ⇒ 영가설 기각     
  (2) 기각역에 포함되지 않음 ⇒ 영가설 받아들임

##9. **가설검정 정리**
- 표본을 이용하여 모집단에 대한 정보를 추정한 결과는 어떤 경우에도 100% 정확하다고 절대로 말할 수 없다.
- 따라서, 모집단에 대한 모든 주장에 대해 신뢰성을 검증할 필요가 있다.
- 6단계의 가설검정을 통해 모집단에 대한 통계적 주장을 검정한다.

- **주요 용어 정리**
  - **영가설**(귀무가설, $H_0$) : 맞다고 가정하는 가설
  - **대립가설**($H_1$) 
    - 영가설에 대립되는 가설    
    - 내용과 상황에 따라 대립가설이 결정됨
    - 기각역 결정에 절대적으로 영향을 끼친다.
  - **유의수준**
    - 참인 영가설을 기각하는 오류를 범할 확률
    - $유의수준 = 1 - 신뢰수준$
    - 즉, 영가설에 포함된 주장이 95% 신뢰수준으로 계산된 내용이면서 동시에 참임에도 불구하고, 표본조사에 의해 부정될 확률이 5%임     
    ⇒ 즉, 유의수준 5%
  - **기각역** : 영가설을 기각하기 위한 기준으로 사용되는 값들의 집합
  - **하단측검정**
    - 대립가설이 "<u>⋯보다 작다</u>"의 모양인 경우에 사용
    - 기각역이 **정규분포 그래프 왼쪽**에 위치함
    - 유의수준 5%에서 기각역은 아래와 같다.
  - **상단측검정**
    - 대립가설이 "<u>…보다 크다</u>"의 모양인 경우에 사용
    - 기각역이 **정규분포 그래프 오른쪽**에 위치함
    - 유의수준 5%에서 기각역은 아래와 같다.
  - **양측검정**
    - 대립가설이 "<u>⋯가 아니다</u>"의 모양인 경우에 사용
    - 기각역이 **정규분포 그래프 양쪽**에 위치함
    - 유의수준 5%에서 기각역은 아래와 같다.

````python
# 하단측검정 
import scipy.stats as st
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-3, 3, 200)
y = st.norm(0, 1).pdf(x)   
plt.xticks(np.arange(-2, 3, 2))
plt.gca().axes.yaxis.set_visible(False)
plt.gca().axes.xaxis.set_visible(False)
cum = np.arange(-3, -1.8, 0.01)
plt.fill_between(cum, st.norm.pdf(cum), color='r')
plt.annotate('5% - 기각역', ha = 'center', xytext = (-2, 0.2), xy = (-2, 0.02), arrowprops = {'facecolor':'k', 'edgecolor':'k'})
plt.plot(x, y)
````

````python
# 상단측검정
x = np.linspace(-3, 3, 200)
y = st.norm(0, 1).pdf(x)   
plt.xticks(np.arange(-2, 3, 2))
plt.gca().axes.yaxis.set_visible(False)
plt.gca().axes.xaxis.set_visible(False)
cum = np.arange(1.8, 3, 0.01)
plt.fill_between(cum, st.norm.pdf(cum), color='r')
plt.annotate('5% - 기각역', ha = 'center', xytext = (2, 0.2), xy = (2, 0.02), arrowprops = {'facecolor':'k', 'edgecolor':'k'})
plt.plot(x, y)
````

````python
# 양측검정
x = np.linspace(-3, 3, 200)
y = st.norm(0, 1).pdf(x)   
plt.xticks(np.arange(-2, 3, 2))
plt.gca().axes.yaxis.set_visible(False)
plt.gca().axes.xaxis.set_visible(False)
cum = np.arange(2, 3, 0.01)
plt.fill_between(cum, st.norm.pdf(cum), color='r')
cum2 = np.arange(-3, -2, 0.01)
plt.fill_between(cum2, st.norm.pdf(cum2), color='r')
plt.annotate('2.5% - 기각역', ha = 'center', xytext = (2.5, 0.2), xy = (2.5, 0.02), arrowprops = {'facecolor':'k', 'edgecolor':'k'})
plt.annotate('2.5% - 기각역', ha = 'center', xytext = (-2.5, 0.2), xy = (-2.5, 0.02), arrowprops = {'facecolor':'k', 'edgecolor':'k'})
plt.plot(x, y)
````
