### 배열


__✏️ 예제__  
: $A, B, C$ 세 마리 말의 경주  

|$x$|-500|3,500|
|:---:|:---:|:---:|
|$P(X=x)$|$\frac{5}{6}$|$\frac{1}{6}$|

▶ $E(x) = -500 × \frac{5}{6} + 3500 × \frac{1}{6} = 168 $

$  $

> $A, B, C$ 세 마리를 순서있게 나열하는 방법이 6가지이기 때문에 분모가 __6__이다.

__< 6가지인 이유 >__  

|1등|2등|3등|
|:---:|:---:|:---:|
|3|2|1|

$3 \times 2 \times 1 = 6$  
$ = 3!$ (factorial, 팩토리얼)

```
1등       A               B               C
         / ＼            / ＼            / ＼
2등     B    C          A    C          A    B
       /      ＼       /      ＼       /      ＼
3등   C         B     C         A     B         A
```

💡 일반화 : $n$개의 대상을 순서대로 나열시킬 수 있는 경우의 수  
$= n!\ (= n × (n-1) × … 2 × 1)$

$  $

__배열이란 ❓__  
- 어떤 사물들의 순서를 따지면서 나열한 것


#### 배열의 응용 ① : 원형배열

- 은행 번호표 등에 사용  

__✏️ 예)__  
$A, B, C, D$ 네 마리의 말을 원형으로 나열하는 방법의 경우의 수는 ? 

```python
A, B, C, D 와 ┐        ┏   A        B
B, C, D, A는  │ 하지만 ┃ B   C 와 A   C 는 # 동일한 원형 배열이다.
#다르다      ┘        └   D         D
```

---

이와 같이, 예를 들어  
A, B, C, D ┐  
B, C, D, A │ 네개의 원형배열로  
C, D, A, B │ 다룰 경우 동일하게 취급된다.  
D, A, B, C ┘  

▶ 이런 방식으로 생각하면 $4!$개의 서로 다른 배열을 4개씩 그룹으로 엮을 수 있다.  
따라서 총 **$\frac{4!}{4}$ 개의 그룹**이 생성되며, 이것이 **4마리를 대상**으로 하는 원형배열의 경우의 수이다. 

---

💡 일반화 : $n$개의 사물을 순서있게 원형배열로 위치시키는 경우의 수  
= $\frac{n!}{n} = (n-1)!$

#### 배열의 응용 ② : 중복배열

__✏️ 예)__  
6마리의 말이 달리기 시합.
- 3마리 : 얼룩말 $(A1, A2, A3)$
- 3마리 : (일반)말 $(B1, B2, B3)$  
6마리를 순서대로 나열하는 경우의 수 $= 6!$

그런데, __말의 종류를 구분하지 않을 떄의 경우의 수는 ?__ 


예를 들어  
$[A1🏇, B2🐎, A3🏇, B1🐎, B3🐎, A2🏇]$와   
$[A2🏇, B1🐎, A1🏇, B3🐎, B2🐎, A3🏇]$ 는 다른 배열이다.   
하지만, 얼룩말 끼리의 순서도 무시하고, 일반 말의 순서도 무시한다면, 두 배열 모두 동일한 모양을 갖는다.  
즉, 모두 아래 모양을 갖는다.  
[🏇,🐎,🏇,🐎,🐎,🏇]

$  $

결론 : __동일한 종류의 순서를 무시하면 경우의 수는 $\frac{6!}{3!3!}$__
- $3!$은 얼룩말 세 마리를 대상으로 한 배열의 숫자
- $3!$은 일반 말 세 마리를 대상으로 한 배열의 숫자

---

💡 일반화 ① : $n$개의 사물을 순서있게 나열하기
- __조건__ : 동일한 종류의 사물이 $k$개 있다.  
그리고, $k$개의 순서는 무시한다
= $\frac{n!}{k!}$

💡 일반화 ② : $n$개의 사물을 순서있게 나열하기
- __조건__ : 동일한 종류의 사물이 $j$개와 $k$개가 있다.   
그리고, $j$개와 $k$개 각각의 순서는 무시한다
= $\frac{n!}{j!k!}$

#### ✏️배열 활용 예제

헬스클럽의 전화번호 맞출 확률  

① 조건   
- $1,2,…,7$의 숫자로 이루어짐
- 7자리
- 중복 없음  

▶ 맞출 확률 = $\frac{1}{7!}=\frac{1}{5040}=0.0002$

② 추가 조건
- 첫째 세자리는 $1,2,3$으로 이루어짐  

총 경우의 수 = $3!×4!$
- $3! = 1,2,3$의 배열
- $4! = 4,5,6,7의 배열$  

▶ 맞출 확률 $= \frac{1}{3!4!} = \frac{1}{144}$


### 순열

__✏️ 예제__  
: 20마리의 말이 경주할 경우에 $1,2,3$등 맞추기

$  $

__방법 ①__  
1등 경우의 수 × 2등 경우의 수 × 3등 경우의 수  
= 20 × 19 × 18

__방법 ②__  
$\frac{20!}{20-3!}$
- $20-3!$ : 3마리 제외한 나머지의 등수는 관심 없음(중복배열 개념 활용 !)

---

💡 일반화 :  n 개의 사물을 순서있게 나열할 때  
처음 $r$개의 순서를 정하는 경우의 수  
${^nP_r}=\frac{n!}{(n-r)!}$  
- $(n-r)!$ : $r$개를 제외한 나머지의 순서는 관심없음

### 조합

__✏️ 예제__  
: 20마리의 말이 경주할 경우에 $3$등 안에 들 세마리 말 맞추기

$  $
세마리 말을 선택하는 경우의 수. 단, 3마리의 순서는 무시

__방법 ①__  
(1등 경우의 수 × 2등 경우의 수 × 3등 경우의 수) / $3!$ (선택된 세마리의 순서 무시) 
= $\frac{20 × 19 × 18}{3!}$

__방법 ②__  
$\frac{20!}{20-3! ⋅ 3!}$
- $20-3!$ : 3마리 제외한 나머지의 등수는 관심 없음(중복배열 개념 활용 !)
- $3!$ : 선택된 세 마리의 배열의 개수

---

💡 일반화 :  n 개의 사물을 순서있게 나열할 때  
처음에 위치할 $r$마리를 선택하는 방법의 경우의 수  
${^nC_r}=\frac{n!}{(n-r)!⋅r!}$  
- $(n-r)!$ : $r$개를 제외한 나머지의 순서는 관심없음
- $r!$ : 선택된 $r$마리의 순서 무시!

#### ✏️순열과 조합 연습  
12명으로 구성된 농구팀에서 시합에 뛸 5명의 선수를 선발해야 함.

① 시합을 뛸 선수들을 선택하는 방법의 경우의 수는 ?
- 12명에서 5명을 선발해야함.
- 순서 없음  

▶ 따라서 경우의 수는  
$^{12}C_5 = \frac{12!}{7!5!} = 792$


② **3명의 선수가 전문 슈터로 구분되었음. 이 세명의 선수가 동시에 시합에서 뛸 확률은 ?**  
▶ 3명이 함께 뛸 경우의 수  
= 나머지 9명에서 2명을 선발하는 경우의 수  
$^{9}C_2 = \frac{9!}{2!7!} = 36$   

$  $

따라서, 3명의 슈터가 동시에 뛸 확률은 $\frac{36}{792} = \frac{1}{22}$
