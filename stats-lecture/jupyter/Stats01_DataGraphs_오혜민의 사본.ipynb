## **7. 히스토그램**

#### ✅**데이터 종류 두 가지**
  1. **범주적 데이터**(질적 데이터)
    - 범주의 성질이나 특성에 따라 나뉘어진 데이터
    - 히스토그램으로 나타낼 수 없음
  2. **수치적 데이터**(양적 데이터)
    - 무게, 길이, 시간, 점수 등 숫자의 범위에 따라 나뉘어진 데이터
    
####✍__예제__ : 시험 점수 구간별 수험생 숫자  

|점수|도수 (학생 수)|
|:------:|:---:|
|0 ~ 199|5|
|200 ~ 399|29|
|400 ~ 599|56|
|600 ~ 799|17|
|800 ~ 999|3|
|✋ 점수 : 범주(점수 구간 활용)|

✋ 막대그래프 활용은 적절치 않음 : 구간별 연속성이 제대로 반영되지 못함



```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(5)
scores = ['0~199', '200~399', '400~599', '600~799', '800~999']
students = [5, 29, 56, 17, 3]

plt.bar(x, students,width=0.6)
plt.xticks(x, scores)
plt.xlabel('scores')
plt.ylabel('students')
plt.show()
```

```python
import matplotlib.pyplot as plt
import random
import numpy as np

a = random.sample(range(800,999),3)
b = random.sample(range(600,799),17)
c = random.sample(range(400,599),56)
d = random.sample(range(200,399),29)
f = random.sample(range(0,199),5)
scores = a+b+c+d+f

# 히스토그램에서 구간 소수점 표기 방법?
plt.hist(scores, bins = np.arange(0,1000,199.5))
plt.axis([0, 1000, 0, 60])
plt.show()
```

✋ 구간 구분을 199, 399, ... 등 대신 199.5, 399.5, ... 등으로 하는 이유는 모든 점수가 하나의 구간에 포함되도록 하기 위함

#### ✅__도수밀도__
- 히스토그램에 사용되는 막대들의 폭은 적절하게 지정하면 된다
- 하지만 구간별 크기가 다를 경우 막대의 높이는 도수의 크기에 비례하도록 해야한다.  
  이를 위해 도수밀도를 계산하여 막대의 높이로 사용해야한다.

  $$도수밀도 = \frac{도수}{구간의 크기} = \frac{도수}{막대의 폭}= 막대의 높이$$


#### ✍ __예제__ : 사람들의 하루 게임시간
|게임시간|도수|
|:------:|:---:|
|0 ~ 1|4,300|
|1 ~ 3|6,900|
|3 ~ 5|4,900|
|5 ~ 10|2,000|
|10 ~ 24|2,100|

✋ 도수 밀도를 사용하지 않으면 상대적으로 적은 도수지만 막대의 면적이 가장 크게 보여지면서 오해를 불러일으킬 수 있다.

```python
import matplotlib.pyplot as plt
import random
import numpy as np

a = (np.random.uniform(0,1,4300)).tolist()
b = (np.random.uniform(1,3,6900)).tolist()
c = (np.random.uniform(3,5,4900)).tolist()
d = (np.random.uniform(5,10,2000)).tolist()
f = (np.random.uniform(10,24,2100)).tolist()
scores = a+b+c+d+f
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(1, 1, 1)

ax.hist(scores, bins = np.arange(-1,24,2))


ax.axis([0, 24, 0, 7000])
ax.set_xticks([0,5,10,20])
```

✋ **해결책** : 도수밀도를 막대의 높이로 사용해야한다. 
            즉, **막대의 넓이가 도수와 일치해야한다**.


|게임시간|도수|구간 크기|도수밀도|
|:------:|:---:|:---:|:---:|
|0 ~ 1|4,300|1|4,300|
|1 ~ 3|6,900|2|3,450|
|3 ~ 5|4,900|2|2,450|
|5 ~ 10|2,000|5|400|
|10 ~ 24|2,100|10|150|

```python
import matplotlib.pyplot as plt
import random
import numpy as np

a = (np.random.uniform(0,1,4300)).tolist()
b = (np.random.uniform(1,3,3450)).tolist()
c = (np.random.uniform(3,5,2450)).tolist()
d = (np.random.uniform(5,10,400)).tolist()
f = (np.random.uniform(10,24,150)).tolist()
scores = a+b+c+d+f

plt.hist(scores, bins = np.arange(-1,24,2))
plt.axis([0, 24, 0, 7000])
plt.show()
```

## 8. 선그래프
- 막대 그래프의 상단 중심부를 선분으로 연결하여 각 범주를 비교하는 그래프

#### __✅특징__
1. 시계열 데이터의 변화 정보를 전달할 때 유용함
2. 구간별 도수의 부분합인 누적 도수는 선 그래프를 활용해야 정확안 정보를 전달할 수 있음
3. 복수의 데이터들의 변화를 한 눈에 전달하고자 할 때 막대 그래프보다 효과적임

#### ✅주요개념 : __시계열 데이터__
- 주가, 지지도, 센서 데이터, 기록 추세 데이터 등 시간별로 구성된 데이터
- 선그래프의 일종인 시계열 그래프에 적합한 데이터

#### ✍ __예제__ 3 : A, B 두 회사의 연도별 수익 비교
- 복수의 시계열 데이터 비교
- 막대그래프처럼 선그래프를 이용하여 복수의 데이터를 하나의 그래프로 나타낼 수 있음

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20,5))

# 막대그래프
plt.subplot(1, 2, 1) 
X=[1,3,5,7,10]
data1 = [1,2,3,4,10]
p1 = plt.bar(X, data1,color='r',width=0.5)

X1 = [X[i] + 0.5 for i in range(len(X))]
data2 = [2,3,4,5,6]
p2 = plt.bar(X1, data2,color='g',width=0.5)

label = ['2007', '2008', '2009', '2010','2011']
plt.ylabel('revenue', fontsize=10)
plt.xlabel('year', fontsize=10)
plt.xticks(X1, label, fontsize=10)
plt.legend((p2[0], p1[0]), ('A', 'B'), fontsize=10)

# 선그래프
plt.subplot(1, 2, 2)  
x = [1,3,5,7,10]

y1 = [1,2,3,4,10]
p1 = plt.plot(x,y1,marker="o",color = "red")

y2 = [2,3,4,5,6]
p2 = plt.plot(x,y2,marker="o",color = "green")

label = ['2007', '2008', '2009', '2010','2011']
plt.ylabel('revenue', fontsize=10)
plt.xlabel('year', fontsize=10)
plt.xticks(X1, label, fontsize=10)
plt.legend((p2[0], p1[0]), ('A', 'B'), fontsize=10)
```
