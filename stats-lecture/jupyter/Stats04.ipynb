### 4. 확률 구하기 일반화
- 사건 $A$가 일어날 확률 : $P(A)$  
$P(A) = \frac{사건\ A가\ 일어날\ 경우의\ 수}{전체\ 경우의\ 수} = \frac{n(A)}{n(S)}$
  - $S$ : 표본공간(모든 가능한 사건들의 공간)
- __주의사항__  
  - 표본공간 $S$의 정확한 의미는 상황과 조건에 의존
  - 나중에 __조건부확률__을 배울 때에 좀 더 자세히 설명될 것임

### 5. 벤 다이어그램 활용

- 표본공간 $S$와 사건 $A$와의 관계를 벤 다이어그램을 이용해 나타낼 수 있음.

```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2

plt.title("S")

v = venn2(subsets = {'10': 1, '01': 0, '11': 0}, set_labels = ('A', 'A`'))
v.get_label_by_id('10').set_text('')
v.get_label_by_id('01').set_text('')

plt.axis('on')
plt.show()
```

- $A$의 여사건$(A')$ : $A$가 일어나지 않을 사건  
$P(A') = 1- P(A)$

- $P(38) = 0$

### 6. 두 사건의 합과 곱
- 질문
  - 사건 $A$ 또는 사건 $B$가 발생할 확률은?
  - 사건 $A$와 사건 $B$가 동시에 발생할 확률은?


- 예 ①
  - $A$ : 검정색
  - $B$ : 빨간색

  $P(A\ 또는\ B) = \frac{36}{38} = \frac{18}{38} + \frac{18}{38} = P(A) + P(B)$

  🔽 $A$와 $B$는 __서로 배반__임 (__두 사건이 동시에 발생하지 않음__)

```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2
plt.rc("font", family='NanumGothic')

plt.title("S")

v = venn2(subsets = {'10': 18, '01': 18, '11': 0},
          set_labels = ('검정색', '빨간색'),
          set_colors=("black", "red"),alpha=0.5)

plt.axis('on')
plt.show()
```

- 예 ②
  - $A$ : 검정색
  - $B$ : 짝수

  $P(A\ 또는\ B) = \frac{26}{38} ≠ \frac{18}{38} + \frac{18}{38} = P(A) + P(B)$

  🔽 $A$와 $B$는 __서로 배반 아님__ (__두 사건이 동시에 발생하는 경우 존재__)

```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2
plt.rc("font", family='NanumGothic')

plt.title("S")

v = venn2(subsets = {'10': 8, '01': 8, '11': 10},
          set_labels = ('검정색', '짝수'),
          set_colors=("black", "green"),alpha=0.5)

plt.axis('on')
plt.show()
```

__사건의 합 확률 계산__  
$P(A\ 또는\ B) = P(A) + P(B) - P(A\ 이고\ B)$  

$  $

▶ 예 ②의 확률  
$P(A\ 또는\ B) = \frac{18}{38} + \frac{18}{38} - \frac{10}{38} = \frac{26}{38}$  
- 이유 : $P(A\ 이고\ B) = \frac{10}{38}$ (룰렛 게임 그림에서 직접 확인 가능)  

---

__용어의 기호화__   
$P(A \cup B) = P(A) + P(B) - P(A∩B)$
- $∪$ : 합집합 기호 ( 또는, 혹은)
- $∩$ : 교집합 기호 ( 이고, 그리고)

🛑 $A$와 $B$가 서로 배반사건인 경우 
 $P(A∩B)=0$ 이다.  
 따라서, $P(A∩B) = P(A) + P(B)$
 - $A∩B$는 중첩되어서 두번 더해진다.


```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2
plt.rc("font", family='NanumGothic')

plt.title("S")

v = venn2(subsets = {'10': 8, '01': 8, '11': 10},
          set_labels = ('A', 'B'),
          set_colors=("black", "red"),alpha=0.5)
v.get_label_by_id('10').set_text('')
v.get_label_by_id('01').set_text('')
v.get_label_by_id('11').set_text('A∩B')

plt.axis('on')
plt.show()
```

### 8. 조건부확률
$P(A|B)$ : 특정 사건 $B$가 발생한다는 조건 하에서 다른 사건 $A$가 발생할 확률

__✏️ 예제__ : 룰렛 게임에서 공이 검정색 구멍에 들어간다는 전제 하에 짝수 구멍에 들어갈 확률은 ?
  - 사건 $A$ : 공이 짝수 구멍에 들어간다.  
  - 사건 $B$ : 공이 검정색 구멍에 들어간다. 

▶ 답 : $P(A|B) = \frac{10(짝수\ 구멍의\ 개수)}{18(검정색\ 구멍의\ 개수}$  

  __주의__
> 분모에 사용되는 표본 공간이 달라졌다.  
  __따라서, $P(A|B) ≠ P(A \cap B)$__ 

$  $

__일반적으로 아래의 식이 성립한다.__  
$P(A|B) = \frac{P(A∩B)}{P(B)}$  
이유 : $P(A|B) = \frac{n(A∩B)}{n(B)} = \frac{\frac{n(A \cap B)}{n(S)}}{\frac{n(B)}{n(S)}} = \frac{P(A∩B)}{P(B)}$ 


__벤 다이어그램으로 보는 조건부 확률__  
🔽 $P(A|B)$를 계산할 때 $B$가 표본공간이 됨. 즉, $S$가 아님.

```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2,venn2_circles
plt.rc("font", family='NanumGothic')

plt.title("S")

v = venn2(subsets = {'10': 8, '01': 8, '11': 10},
          set_labels = ('A', 'B'),
          set_colors=("white", "purple"),alpha=0.5)
v.get_label_by_id('10').set_text('')
v.get_label_by_id('01').set_text('')
v.get_label_by_id('11').set_text('')

c = venn2_circles(subsets = {'10': 8, '01': 8, '11': 10},
                  linewidth=1)


plt.axis('on')
plt.show()
```

### 9. 베이즈 정리
- 조건부 확률 계산을 도움
- 불확실한 상황에서 의사결정 문제를 수학적으로 다룰 때 매우 중요한 도구 제공
- 주어진 확률을 이용하여 미래의 확률을 계산할 때 활용


$P(A|B) = \frac{P(A) \times P(B|A)}{P(A)×P(B|A)\ +\ P(A')×(P(B|A')}$

__✏️ 예__ : 게임 회사에서 자체 개발한 게임 두 개를 테스트하기 위해 지원자를 모집하여 아래의 결과를 얻었음.
- 테스트 지원자 중 80%는 게임1 선택
- 테스트 지원자 중 20%는 게임2 선택
- 게임1을 테스트한 사람 중에 60%가 만족
- 게임 2를 테스트한 사람 중에 70%가 만족

$  $

__질문__ : 테스트 지원자 중에 임의로 한 사람을 선택하여 게임 만족 여부를 물었더니 그렇다고 하였음. 이때 그 사람이 게임 2를 선택했었을 확률은 ?

🔑 힌트 : 베이즈 정리 활용

####  

__해법__ : 베이즈 정리를 활용하기 위해 두 개의 사건 $A$와 $B$를 정의해야함.
- 사건 $A$ : 게임2 선택
- 사건 $B$ : 게임에 만족

▶ $P(A|B)$를 계산해야함.  
먼저 아래 사실을 확인할 수 있다.
- $A'$
- $P(A')=0.8$
- $P(A)=0.2$
- $P(B|A')=0.6$
- $P(B|A)=0.7$

이제 베이즈 정리를 이용할 수 있다.  
$P(B|A) = \frac{P(A)×P(B|A)}{P(A)×P(B|A)+P(A')×P(B|A')} = \frac{0.2×0.7}{0.2×0.7+0.8×0.6} = \frac{0.14}{0.62} = \frac{7}{31} =0.226$

__✏️ 연습문제__  
어느 병원에서 암진단에 대한 아래의 자료를 갖고있다.
- 암진단 검사를 받은 사람 중에 암에 걸린 확률 : 20% 
- 암에 걸린 사람 중에 암진단 결과가 양성인 확률 : 90%
- 암에 걸리지 않았지만 암진단 결과가 양성인 확률 : 30%

__질문__ : 암진단 결과가 양성이지만 실제로는 암이 아닐 확률 ? 
####  

__해법__  
베이즈 정리를 이용하기 위해 사건 $A$와 $B$를 다음과 같이 정의한다.
- 사건 $A$ : 암에 걸리지 않았음
- 사건 $B$ : 암진단 양성

▶ $A'$ : 암에 걸렸음.
- $P(A') = 0.2$

▶$P(A|B) = \frac{P(A)×P(B|A)}{P(A)×P(B|A)+P(A')×P(B|A')} = \frac{0.8×0.3}{0.8×0.3+0.2×0.9} = \frac{0.24}{0.24+0.18} = \frac{24}{42} = \frac{4}{7} = 0.57$

- 이유 : $P(B|A)$가 너무 높음

### 10. 종속사건 대 독립사건
- 두 개의 사건 $A$와 $B$가 서로 영향을 주면 두 사건 사이에 종속관계가 성립된다.
- 반면에 서로 영향을 주지 않으면 서로 독립이라고 말한다.

---
- 종속여부 확인 방법①
  - $A$와 $B$가 서로 종속 ↔ $P(A|B)≠P(A)$
  - $A$와 $B$가 서로 독립 ↔ $P(A|B)= P(A)$  

  __주의__ : $P(A|B)= P(A)$이면 $P(B|A)= P(B)$이다. 

- 종속여부 확인 방법②
  - $A$와 $B$가 서로 종속 ↔ $P(A \cap B)≠P(A)×P(B)$
  - $A$와 $B$가 서로 독립 ↔ $P(A \cap B)=P(A)×P(B)$

__✏️ 연습문제__  
룰렛게임에서 아래 두 사건의 종속/독립 여부 확인하기
- 사건 $A$ : 짝수 구멍에 멈춘다.
- 사건 $B$ : 검정색 구멍에 멈춘다.

🔑 힌트 : 공을 한 번 굴리는 경우와 두 번 굴리는 경우에 따라 종속/독립 여부 달라짐

####  

__해답__  
경우 ① : 공을 한 번 굴리는 경우  
$P(짝수|검정색) = 공이\ 검정색\ 구멍에\ 들어갔다는\ 전제\ 하에\ 짝수\ 구멍일\ 확률 = \frac{10}{18}≠\frac{18}{38}= P(짝수)$  
따라서, 사건 $A$와 사건 $B$는 서로 종속이다.

$  $

경우 ② : 공을 두 번 굴리는 경우  
$P(짝수|검정색) = 첫\ 째\ 공이\ 검정색\ 주머니에\ 들어갔다는\ 전제\ 하에\ 두\ 번째\ 공이\ 짝수\ 주머니에\ 들어갈\ 확률 = \frac{10}{18}= 0.474= P(짝수)$    
따라서, 사건 $A$와 사건 $B$는 서로 독립이다.

__✏️ 연습문제__  
열 장의 카드가 있다. 그중에 다섯 장은 검은색이고 나머지 다섯 장은 빨간 색이다.  
이제 카드 두 장을 한 장씩 임의로 선택한다.  
아래 두 사건의 종속/독립 여부를 판단하라.
- 사건 $A$ : 뽑힌 카드는 검정색이다.
- 사건 $B$ : 뽑힌 카드는 빨간색이다.
####  

__해답__  
카드를 뽑는 방식에 따라 종속/독립 여부가 달라진다.
- 먼저, $P(A) = P(B) = \frac{5}{10} = \frac{1}{2}$임을 확인한다.

$  $

▶ 방식 ①  
: 뽑은 카드를 다시 넣고 섞는 경우  
$P(A|B) = \frac{5}{10} = P(A)$  
그리고  
$P(A \cap B) = \frac{5}{10} × \frac{5}{10} = P(A) × P(B)$  

▶ 방식 ②  
: 뽑은 카드는 제외하고 다시 뽑을 경우  
$P(A|B) = \frac{5}{9} ≠ \frac{5}{10} = P(A)$  
그리고  
$P(A \cap B) = \frac{5}{10} × \frac{5}{9} ≠ \frac{5}{10} × \frac{5}{10}= P(A) × P(B)$
