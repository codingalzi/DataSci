# 12장. 신뢰수준(2부)

### 1. 점추정의 문제와 해결책

##### 1) 문제 
__①__ 모집단의 평균과 분산을 위해 표본을 활용한 점추정 기법을 살펴보았다. 그런데 추정값을 얻기 위해 단 하나의 표본만을 이용하였다.

__②__ 표본이 편향되지 않았다 하더라도 표본이 모집단을 100% 정확하게 반영하는지 여부는 절대로 알 수 없다.

__③__ 점추정은 단지 최선을 예측하는 것 뿐이다.

##### 2) 해결책
__①__ 예를 들어, 비율에 대한 추정값을 명확하게 말하는 대신에 비율의 추정값이 어느 구간에 속하는지를 말할 수 있다.

__②__ 구간의 크기는 결과를 어느 정도 신뢰할 수 있게 만들 것인가에 따라 결정된다.

__③__ 신뢰할 수 있는 정도를 __신뢰수준__이라 부름.

### 2. 점추정 신뢰구간 구하기 단계 정리
__①__ 모집단 통계 선택
- 예) 빨간색 풍선껌의 비율에 대한 신뢰구간 설정 필요.  
즉, $P$에 대한 신뢰구간 설정 필요  

__②__ 표본 분포 찾기
- 예 ) 비율 표본 분포의 기대치와 분산 계산  
$P_s$의 정규분포 활용  

__③__ 신뢰 수준 정하기  
- 예 ) 일반적으로 95%의 신뢰수준 사용

__④__ 신뢰구간 찾기 
- 예) 신뢰수준과 표본 분포를 사용하여 신뢰구간 구하기


### 3. 신뢰구간 설정 공식 모음

__①__
- 모집단 통계 : μ  
- 모집단 분포 : $X$ 정규분포  
- 조건 : 
  - $σ^2$ 알고있음  
  $n$ 표본 크기  
  X̄표본 평균 값
- 신뢰구간 : $($X̄$-c\frac{σ}{\sqrt{n}},\ $X̄$+c\frac{σ}{\sqrt{n}})$  

__②__
- 모집단 통계 : μ  
- 모집단 분포 : $X$ 정규분포 아님
- 조건 : 
  - $σ^2$ 알고있음  
  $n \ge 30$  
  X̄표본 평균 값
- 신뢰구간 : $($X̄$-c\frac{σ}{\sqrt{n}},\ $X̄$+c\frac{σ}{\sqrt{n}})$  

__③__
- 모집단 통계 : μ  
- 모집단 분포 : $X$ 임의의 분포  
- 조건 : 
  - $σ^2$ 모름  
  $n \ge 30$  
  X̄표본 평균 값
  $S^2$표본 분산
- 신뢰구간 : $($X̄$-c\frac{s}{\sqrt{n}},\ $X̄$+c\frac{s}{\sqrt{n}})$  

__④__  
- 모집단 통계 : $P$
- 모집단 분포 : 이항  
- 조건 : 
  - $n \ge 30$  
  $P_s$ 표본비율  
  $1_s = 1-P_s$
- 신뢰구간 : $(P_s-c\sqrt{\frac{P_s⋅q_s}{n}},\ P+c\sqrt{\frac{P_s⋅q_s}{n}})$
  - $c\sqrt{\frac{P_s⋅q_s}{n}}$ : 오차범위 






▶ c의 값 : 신뢰 수준에 의해 결정됨

|신뢰수준|c의 값|
|:---:|:---:|
|90%|1.64|
|95%|1.96|
|99%|2.58|

아래 그림에서 표집 오차 부분을 이해할 수 있어야한다.  

image.png

- 총 응답자 : 2,523명  
신뢰수준 : 95%  
오차범위 : ±2.0%

__신뢰오차 ±2.0%에 이용된 "2.0%"에 대한 근거__  
__①__ 신뢰수준 = 95% ➡ $c = 1.96$  
__②__ $P_s = 0.677,\ q_s = 1-P_s = 0.323$  
__③__ $n=2,523$  
  
➡ $c⋅\sqrt{\frac{P_s⋅q_s}{n}} = 1.96×\sqrt{\frac{0.677×0.323}{2523}} = 0.018$ (약 2%)
